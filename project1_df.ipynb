```python
df['Purchase Date']=pd.to_datetime(df['Purchase Date'])
```

```python
df['Purchase Year']=df['Purchase Date'].dt.year
df['Purchase Month']=df['Purchase Date'].dt.month
df['Purchase day of week']=df['Purchase Date'].dt.dayofweek
df['Purchase day']=df['Purchase Date'].dt.day
df['Purchase hour']=df['Purchase Date'].dt.hour
df['Purchase quarter']=df['Purchase Date'].dt.quarter
```

# EDA

```python
df.columns
```

```python
#YEARLY SALES ANALYSIS
yearly_sales=df.groupby(['Purchase Year'],as_index=False)['Net Amount'].sum()
yearly_sales
sns.lineplot(data=yearly_sales,x='Purchase Year',y='Net Amount',marker='*',markersize=10,color='purple')
plt.show()
```

```python
#QUARTERLY SALES ANALYSIS
quarterly_sales=df.groupby(['Purchase quarter'],as_index=False)['Net Amount'].sum()
print(quarterly_sales)
sns.lineplot(data=quarterly_sales,x='Purchase quarter',y='Net Amount',marker='*',markersize=10,color='blue')
plt.show()
```

```python
#WEEKLYSALES ANALYSIS
weekly_sales=df.groupby(['Purchase day of week'],as_index=False)['Net Amount'].sum()
print(weekly_sales)
sns.lineplot(data=weekly_sales,x='Purchase day of week',y='Net Amount',marker='*',markersize=10,color='red')
plt.show()
```

```python
#HOURLYSALES ANALYSIS
hourly_sales=df.groupby(['Purchase hour'],as_index=False)['Net Amount'].sum()
print(hourly_sales)
sns.barplot(data=hourly_sales,x='Purchase hour',y='Net Amount',color='green')
plt.show()
```

Not too much difference in houly sales

```python
# gender wise sales
gender=df.groupby(['Gender'],as_index=False)['Net Amount'].sum()
print(gender)
plt.pie(gender['Net Amount'],labels=gender['Gender'],startangle=140,autopct='%1.1f%%')
plt.show()
```

```python
#sales by age group
age_group=df.groupby(['Age Group'],as_index=False)['Net Amount'].sum()
print(age_group)
sns.barplot(data=age_group,x='Age Group',y='Net Amount')
plt.show()
```

```python
# sales by location
loc=df.groupby(['Location'],as_index=False)['Net Amount'].sum().sort_values(by='Net Amount',ascending=False)
print(loc)
sns.barplot(data=loc,x='Location',y='Net Amount',palette={'blue','lightgreen','red','yellow','red'})
plt.xticks(rotation=75)
plt.show()
```

```python
#sales by product category
prod_cat=df.groupby(['Product Category'],as_index=False)['Net Amount'].sum().sort_values(by='Net Amount',ascending=False)
print(prod_cat)
sns.barplot(data=prod_cat,x='Product Category',y='Net Amount',palette={'lightblue','yellow','red'})
plt.xticks(rotation=75)
plt.show()
```
